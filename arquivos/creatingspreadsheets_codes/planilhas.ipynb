```python
tcia_excel = 'tcia-diagnosis-data-2012-04-20.xls'
lidc_excel = 'lidc-idri-nodule-counts-(6-23-2015).xlsx'

df1 = pd.read_excel(tcia_excel)
df1
```

```python
df2 = pd.read_excel(lidc_excel)
df2
```

```python
def rename_column(df, column, new_name):
    df.rename(columns={column: new_name}, inplace=True)

rename_column(df1, 'TCIA Patient ID', 'patient_id')

rename_column(df1, 'Diagnosis at the Patient Level\n0=Unknown\n1=benign or non-malignant disease\n2= malignant, primary lung cancer\n3 = malignant metastatic\n',
              'patient_diagnosis')
rename_column(df1, 'Diagnosis Method\n0 = unknown\n1 = review of radiological images to show 2 years of stable nodule\n2 = biopsy\n3 = surgical resection\n4 = progression or response',
              'patient_diagnosis_method')

rename_column(df1, 'Primary tumor site for metastatic disease', 'tumor_site')

rename_column(df1, 'Nodule 1\nDiagnosis at the Nodule Level \n0=Unknown\n1=benign or non-malignant disease\n2= malignant, primary lung cancer\n3 = malignant metastatic)\n',
              'nodule1_diagnosis')
rename_column(df1, 'Nodule 1\nDiagnosis Method at the Nodule Level\n0 = unknown\n1 = review of radiological images to show 2 years of stable nodule\n2 = biopsy\n3 = surgical resection\n4 = progression or response\n',
              'nodule1_diagnosis_method')

rename_column(df1, 'Nodule 2\nDiagnosis at the Nodule Level \n0=Unknown\n1=benign or non-malignant disease\n2= malignant, primary lung cancer\n3 = malignant metastatic)\n',
              'nodule2_diagnosis')
rename_column(df1, 'Nodule 2\nDiagnosis Method at the Nodule Level\n0 = unknown\n1 = review of radiological images to show 2 years of stable nodule\n2 = biopsy\n3 = surgical resection\n4 = progression or response\n',
              'nodule2_diagnosis_method')

rename_column(df1, 'Nodule 3\nDiagnosis at the Nodule Level \n0=Unknown\n1=benign or non-malignant disease\n2= malignant, primary lung cancer\n3 = malignant metastatic)\n',
              'nodule3_diagnosis')
rename_column(df1, 'Nodule 3\nDiagnosis Method at the Nodule Level\n0 = unknown\n1 = review of radiological images to show 2 years of stable nodule\n2 = biopsy\n3 = surgical resection\n4 = progression or response\n',
              'nodule3_diagnosis_method')

rename_column(df1, 'Nodule 4\nDiagnosis at the Nodule Level \n0=Unknown\n1=benign or non-malignant disease\n2= malignant, primary lung cancer\n3 = malignant metastatic)\n',
              'nodule4_diagnosis')
rename_column(df1, 'Nodule 4\nDiagnosis Method at the Nodule Level\n0 = unknown\n1 = review of radiological images to show 2 years of stable nodule\n2 = biopsy\n3 = surgical resection\n4 = progression or response\n',
              'nodule4_diagnosis_method')

rename_column(df1, 'Nodule 5\nDiagnosis at the Nodule Level \n0=Unknown\n1=benign or non-malignant disease\n2= malignant, primary lung cancer\n3 = malignant metastatic)\n',
              'nodule5_diagnosis')
rename_column(df1, 'Nodule 5\nDiagnosis Method at the Nodule Level\n0 = unknown\n1 = review of radiological images to show 2 years of stable nodule\n2 = biopsy\n3 = surgical resection\n4 = progression or response\n',
              'nodule5_diagnosis_method')

df1
```
